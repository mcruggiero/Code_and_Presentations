### Human Demonstration

```python
import sys

env = gym.make('CartPole-v1')

if not hasattr(env.action_space, 'n'):
    raise Exception('Keyboard agent only supports discrete action spaces')
ACTIONS = env.action_space.n
SKIP_CONTROL = 0    # Use previous control decision SKIP_CONTROL times, that's how you
                    # can test what skip is still usable.

human_agent_action = 0
human_wants_restart = False
human_sets_pause = False

def key_press(key, mod):
    global human_agent_action, human_wants_restart, human_sets_pause
    if key==0xff0d: human_wants_restart = True
    
    111#Key 32 is the space key
    if key==32: human_sets_pause = not human_sets_pause
    a = int( key - ord('0') )
    if a <= 0 or a >= ACTIONS: return
    human_agent_action = a

def key_release(key, mod):
    global human_agent_action
    a = int( key - ord('0') )
    if a <= 0 or a >= ACTIONS: return
    if human_agent_action == a:
        human_agent_action = 0

env.render()
env.unwrapped.viewer.window.on_key_press = key_press
env.unwrapped.viewer.window.on_key_release = key_release

def rollout(env):
    global human_agent_action, human_wants_restart, human_sets_pause
    human_wants_restart = False
    obser = env.reset()
    skip = 0
    total_reward = 0
    total_timesteps = 0
    while 1:
        if not skip:
            print("taking action {}".format(human_agent_action))
            print(obser)
            a = human_agent_action
            total_timesteps += 1
            skip = SKIP_CONTROL
        else:
            skip -= 1

        obser, r, done, info = env.step(a)
        #if r != 0:
            #print("reward %0.3f" % r)
        total_reward += r
        window_still_open = env.render()
        if window_still_open==False: return False
        if done: break
        if human_wants_restart: break
        while human_sets_pause:
            env.render()
            sleep(0.1)
        sleep(0.1)
    print("timesteps %i reward %0.2f" % (total_timesteps, total_reward))

rollout(env)
env.close()
env.env.close()
```

# Functions and Classes

### Report Function

```python
def report(report_interval,
           episode,
           start_time,
           steps_total,
           epsilon,
           frames_total,
           solved):
    
    if episode % report_interval == 0 and episode != 0:
        print("""\t\t\t *** Episode {} at Total Seconds {:2f}***
                 Average Reward for last {} steps: {:2f}
                 Average Reward for all steps: {:2f}
                 Epsilon: {}, Frames Total :{}
                 Solved {} Times
              """.format(episode,
                   (time() - start_time),
                   str(report_interval),
                   sum(steps_total[-report_interval:])/report_interval,
                   sum(steps_total)/len(steps_total),
                   epsilon,
                   frames_total,
                   solved))
```

## Agent Factory 

```python
class Agent(object):
  
    def factory(type, object):
        if type == "Random":
            return Random_Agent(type, object)
        elif type == "OLN":
            return OLN(type, object)
        elif type == "TLN":
            return TLN(type, object)
        elif type == "ERN":
            return ERN(type, object)
        elif type == "DQN":
            return DQN(type, object)
        elif type == "DoubleDQN":
            return DoubleDQN(type, object)
        elif type == "DuelingDQN":
            return DuelingDQN(type, object)
        elif type == "Atari":
            return Atari(type, object)
        
    factory = staticmethod(factory)
```

### Simulation Function

```python
mem = False
def simulation(num_episodes,
               agent,
               label,
               record,
               report_interval,
               extra_name  = "",
               show = False,
               cart = True):
    
    #Simulation Initialization
    actor = Agent.factory(agent, label)
    start_time = time()
    solved, frames_total = 0, 0
    steps_total = []
    
    #if extra_name != "": extra_name = "_" + extra_name
    
    #The recorded dictionary for the simulation
    shadow_run = {}
    
    #Loop through different episodes
    for episode in range(num_episodes):

        # Episode Initializations
        state = env.reset()
        epi_name = actor.name + extra_name + "_"+ str(episode).zfill(4)
        shadow_run[epi_name] = {} 
        
        #A fancy way to make lists
        blanks = [-1 for i in range(500)]
        
        #Input for episode
        #actions, reported_values = [-1 for i in range(500)], [-1 for i in range(500)]
        actions, reported_values = [], []

        step = 0
                
        while True:
            #This allows us to watch the model in action
            if show == True: rendered = env.render(mode = 'rgb_array')

            
            #Calculate E-greedy Rate for Exploration vs. Exploitation tradeoff
            epsilon = actor.calculate_epsilon(frames_total)

            #Actor acts
            action = actor.select_action(state, epsilon)

            #Record States
            new_state, reward, done, info = env.step(action)

            #Optimize
            #Memory Push check
            
            if mem != False:
                memory.push(state, action, new_state, reward, done)
                actor.optimize()
            else:
                actor.optimize(state, action, new_state, reward, done)

            #Store Inputs for step
            actions.append(action)
            reported_values.append(list(new_state))

            if done:
                steps_total.append(step)
                
                ###
                #**DO THIS**
                #On update, clean up the following syntax
                #Also, consider adding number values to the information
                ###
                
                #Below is an overpacked syntax to add a column for each part
                #.zfill pads zeros in the front of the string
                
                for i in range(500):
                    D = shadow_run[epi_name]
                    #D["action_{}".format(str(i).zfill(3))] = actions[i]                                   
                
                D["agent"] = agent 
                D["steps_total"] = step
#                 D["actions_total"] = actions
                D["outputs_total"] = reported_values
                D["all_actions"] = actions
                shadow_run[epi_name] = D
                #Not needed for this environment
                #shadow_run[episode]["information"] = information

                if step >= score_to_solve: solved += 1

                report(report_interval, episode, start_time, steps_total, epsilon, frames_total, solved)

                break

            step += 1
            frames_total += 1
            state = new_state
        
    #Clean up after Episodes
    record[actor.name + extra_name] = shadow_run    
    env.close()
    env.env.close()

    return record
```

## Q Learning Functions and Methods

### Calculate Epsilon

<strong>The following is part of fundamental algorithm used throughout this notebook. Ultimately, our agents need to manage the Exploitation vs Exploration problem. The following solution uses an <i>E-greedy</i> algorithm that makes short term sacrifices to gather information, and then, once the training phase has ended, exploits its previous experience, deviating from the optimal strategy only rarely. <br><br><i>calculate_epsilon</i> makes the exploration threshold and <i>select_action<i> decides whether to apply the found strategy or choose randomly.</strong>

```python
#We have to add self since these will become methods
def calculate_epsilon(self, steps_done):
    #http://home.deib.polimi.it/restelli/MyWebSite/pdf/rl5.pdf
    #Exploitation vs Exploration algorithm
    epsilon = egreedy_final + (egreedy - egreedy_final) * \
              math.exp(-1. * steps_done / egreedy_decay )
    return epsilon
```

### Select Action

```python
def select_action(self,state,epsilon):

    random_for_egreedy = torch.rand(1)[0]

    #If random > epsilon, then will not attempt random action
    if random_for_egreedy > epsilon:      
        with torch.no_grad():
            #The idea is that we let network choose what it thinks is the best action
            state = Tensor(state).to(device)
            action_from_nn = self.nn(state)
            action = torch.max(action_from_nn,0)[1]
            action = action.item()

    #Otherwise random action is attempted and recorded
    else:
        action = env.action_space.sample()

    return action
```

### Optimize

<strong>The <i>optimize</i> function is the most changed throughout this notebook. Updates often include running experience through a Tensor, avoiding the batching problem (discussed later) and using a learning network and action network.</strong><br><br>

$$Q(s,a)=r+\gamma*\max_{a^{\prime}}Q\left(s^{\prime}, a^{\prime}\right)$$
<br><br>
<strong>The basic point here is that r is our reward, s is our state, and gamma is our discount rate from past actions</strong>


```python
def optimize(self, state, action, new_state, reward, done):
    #Train network
    state = Tensor(state).to(device)
    
    #The reason for the detach values here is that, while we need to calculate
    #The state values, we don't want to calculate the gradient for the new
    #State when we back propigate
    new_state = Tensor(new_state).to(device)

    reward = Tensor([reward]).to(device)

    if done:
        #Prevents Divergence
        target_value = reward
    else:
        #Using detach so we don't update the network
        new_state_values = self.nn(new_state).detach()
        
        #The best max state
        max_new_state_values = torch.max(new_state_values)
        
        #The Bellman Equation
        target_value = reward + gamma * max_new_state_values

    #The is where the network calculates the predictions
    predicted_value = self.nn(state)[action]

    #The learing between the loss and actual
    loss = self.loss_func(predicted_value, target_value)

    #Backwards propigate through the netowork
    self.optimizer.zero_grad()
    
    #We are only calculating this inside the function, we don't need self
    loss.backward()
    
    #Agent Learns
    self.optimizer.step()
```

## Experience Replay Network

### TLN - ERN Differences

<br>
<strong>According to the paper <span style="text-decoration:underline">Playing Atari with Deep Reinforcement Learning</span>, two main problems arise from simple neural nets in RL problems. First, there is a delay between action and rewards; these problems are thus "particularly daunting when compared to the direct association between inputs and targets found in supervised learning." Second, "most deep
learning algorithms assume the data samples to be independent, while in reinforcement learning one
typically encounters sequences of highly correlated states." In other words, the data distributions are non stationary.<br><br>
Deep Mind's solution to this problem was to use: "an experience replay mechanism which randomly samples previous transitions, and thereby smooths the training distribution over many past behaviors." These samples are taken in batches, not sequentially.</strong>

<img src="./DQN.png">

<br>
<strong>In terms of our Pytorch Algorithm, This will involve two major improvements:
1. The addition of the Experience Replay Class
1. Pushing all memory inside of tensors</strong>

### Experience Replay Class

```python
class ExperienceReplay(object):
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = []
        self.position = 0
     
    #Now Training is batches to avoid the clustering and oversampling
    #Of past training
    def push(self, state, action, new_state, reward, done):
        transition = (state, action, new_state, reward, done)
        
        #Checks to see if memory ready to be filled
        if self.position >= len(self.memory):
            self.memory.append(transition)
        else:
            #Overwriting exisiting memory if overfilled
            self.memory[self.position] = transition
        
        self.position = ( self.position + 1 ) % self.capacity
        
        
    def sample(self, batch_size):
        #aggregates elements based on iterables, here it is making
        #batches of experiences, using random sampling
        return zip(*random.sample(self.memory, batch_size))
        
    
    #Verify Memory size
    def __len__(self):
        return len(self.memory)
```

### New Optimize Function

<strong>The ExperienceReplay function is carrying a lot of the sampling transition for learning, thus a significant update is necessary for our optimize function</strong>

```python
#Instead of feeding the optimize function all of the state information
#We will sample if from memory
def mem_optimize(self):

    #If memory is smaller than batch size, we will not run the program
    if (len(memory) < batch_size):
        return

    #Changing the batch_size signifantly effects our outputs.
    state, action, new_state, reward, done = memory.sample(batch_size)

    #Feed in the states into the network
    state = Tensor(state).to(device)
    new_state = Tensor(new_state).to(device)
    reward = Tensor(reward).to(device)
    
    #We are going to need 64-bit integer tensors since we 
    #Are using them as indicies, hence the LongTensor instead of the float
    action = LongTensor(action).to(device)
    done = Tensor(done).to(device)

    new_state_values = self.nn(new_state).detach()
    #Only want the max of the index
    max_new_state_values = torch.max(new_state_values, 1)[0]
    
    #If the episode is done, we are only going to want to reward values
    #Hence, the (1 - done) to zero out the gamma and state max
    target_value = reward + ( 1 - done ) * gamma * max_new_state_values

    #We have to make sure the dimensions match, hence the unsqueeze
    #gather takes the right values from the Neural Net
    predicted_value = self.nn(state).gather(1, action.unsqueeze(1)).squeeze(1)

    loss = self.loss_func(predicted_value, target_value)

    self.optimizer.zero_grad()
    loss.backward()
    self.optimizer.step()
```

```python
def DQN_optimize(self):

    if (len(memory) < batch_size):
        return

    state, action, new_state, reward, done = memory.sample(batch_size)

    state = Tensor(state).to(device)
    new_state = Tensor(new_state).to(device)
    reward = Tensor(reward).to(device)
    action = LongTensor(action).to(device)
    done = Tensor(done).to(device)

    #The New states are updated here, use detach to control updates
    new_state_values = self.target_nn(new_state).detach()
    max_new_state_values = torch.max(new_state_values, 1)[0]
    target_value = reward + ( 1 - done ) * gamma * max_new_state_values

    predicted_value = self.nn(state).gather(1, action.unsqueeze(1)).squeeze(1)

    loss = self.loss_func(predicted_value, target_value)

    self.optimizer.zero_grad()
    loss.backward()

    #This is also an addition from https://storage.googleapis.com/deepmind-media/dqn/DQNNaturePaper.pdf
    #This point is added for completeness
    if clip_error:
        for param in self.nn.parameters():
            param.grad.data.clamp_(-1,1)

    self.optimizer.step()

    if self.update_target_counter % update_target_frequency == 0:
        self.target_nn.load_state_dict(self.nn.state_dict())

    #Counter to control how often the updates occurr
    self.update_target_counter += 1
```

## Double DQN Network

### Double DQN Information

<strong>One problem with our current algorithm is that since we are taking the max expected value for every action, our models tend to overestimate the value of their actions. The solution is to look at the index</strong>
<img src="./DoubleDQNCompare.png">
<img src="./Dcom.png">

<div style="text-align: right">Source: https://arxiv.org/pdf/1509.06461.pdf</div><br><br>
<strong>
<br>
$$
Y_{t}^{\mathrm{DoubleD} \mathrm{Q} \mathrm{N}} \equiv R_{t+1}+\gamma Q\left(S_{t+1}, \underset{a}{\operatorname{argmax}} Q\left(S_{t+1}, a ; \boldsymbol{\theta}_{t}\right), \boldsymbol{\theta}_{t}^{-}\right)
$$
<br><br>The idea is that we set one network (target NN) for calculating the expected value, and another another (learning NN) to choose the best action index. The big point, removing all noise from agent while it trains is impossible, however now we have different types of noise for both, and these errors are no longer correlated to each other!</strong>


### Double DQN Optimize

```python
def DoubleDQN_optimize(self):
        
        if (len(memory) < batch_size):
            return
        
        state, action, new_state, reward, done = memory.sample(batch_size)
        
        state = Tensor(state).to(device)
        new_state = Tensor(new_state).to(device)
        reward = Tensor(reward).to(device)
        action = LongTensor(action).to(device)
        done = Tensor(done).to(device)

        #The big implementation, here we are looking at the correct indexes and then 
        #Running those indices through the analysis. 
        new_state_indexes = self.nn(new_state).detach()
        max_new_state_indexes = torch.max(new_state_indexes, 1)[1]  

        new_state_values = self.target_nn(new_state).detach()

        #Squeeze and Unsqueeze ensure the dimensions fit
        max_new_state_values = new_state_values.gather(1, max_new_state_indexes.unsqueeze(1)).squeeze(1)
        
        target_value = reward + ( 1 - done ) * gamma * max_new_state_values
  
        predicted_value = self.nn(state).gather(1, action.unsqueeze(1)).squeeze(1)
        
        loss = self.loss_func(predicted_value, target_value)
    
        self.optimizer.zero_grad()
        loss.backward()
        
        if clip_error:
            for param in self.nn.parameters():
                param.grad.data.clamp_(-1,1)
        
        self.optimizer.step()
        
        if self.update_target_counter % update_target_frequency == 0:
            self.target_nn.load_state_dict(self.nn.state_dict())
        
        self.update_target_counter += 1
```

### Convolutional Neural Net

```python
class Convolutional_Network(nn.Module):
    def __init__(self):
        super(Convolutional_Network, self).__init__()

        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=8, stride=4)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=4, stride=2)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, stride=1)
        
        self.advantage1 = nn.Linear(7*7*64,hidden_layer)
        self.advantage2 = nn.Linear(hidden_layer, number_of_outputs)
        
        self.value1 = nn.Linear(7*7*64,hidden_layer)
        self.value2 = nn.Linear(hidden_layer,1)

        #self.activation = nn.Tanh()
        self.activation = nn.ReLU()
        
        
    def forward(self, x):
        
        if normalize_image:
            x = x / 255
        
        output_conv = self.conv1(x)
        output_conv = self.activation(output_conv)
        output_conv = self.conv2(output_conv)
        output_conv = self.activation(output_conv)
        output_conv = self.conv3(output_conv)
        output_conv = self.activation(output_conv)
        
        output_conv = output_conv.view(output_conv.size(0), -1) # flatten
        
        output_advantage = self.advantage1(output_conv)
        output_advantage = self.activation(output_advantage)
        output_advantage = self.advantage2(output_advantage)
        
        output_value = self.value1(output_conv)
        output_value = self.activation(output_value)
        output_value = self.value2(output_value)
        
        output_final = output_value + output_advantage - output_advantage.mean()

        return output_final
```

### Calculate Epsilon

```python
def atari_calculate_epsilon(self, steps_done):
    epsilon = egreedy_final + (egreedy - egreedy_final) * \
              math.exp(-1. * steps_done / egreedy_decay )
    return epsilon
```

### Atari Select Action Function

```python
def atari_select_action(self,state,epsilon):

    random_for_egreedy = torch.rand(1)[0]

    if random_for_egreedy > epsilon:      

        with torch.no_grad():

            state = preprocess_frame(state)
            action_from_nn = self.nn(state)

            action = torch.max(action_from_nn,1)[1]
            action = action.item()        
    else:
        action = env.action_space.sample()

    return action
```

### Atari Optimize Function

```python
def atari_optimize(self):

    if (len(memory) < batch_size):
        return

    state, action, new_state, reward, done = memory.sample(batch_size)

    state = [ preprocess_frame(frame) for frame in state ] 
    state = torch.cat(state)

    new_state = [ preprocess_frame(frame) for frame in new_state ] 
    new_state = torch.cat(new_state)

    reward = Tensor(reward).to(device)
    action = LongTensor(action).to(device)
    done = Tensor(done).to(device)


    new_state_indexes = self.nn(new_state).detach()
    max_new_state_indexes = torch.max(new_state_indexes, 1)[1]  

    new_state_values = self.target_nn(new_state).detach()
    max_new_state_values = new_state_values.gather(1, max_new_state_indexes.unsqueeze(1)).squeeze(1)



    target_value = reward + ( 1 - done ) * gamma * max_new_state_values

    predicted_value = self.nn(state).gather(1, action.unsqueeze(1)).squeeze(1)

    loss = self.loss_func(predicted_value, target_value)

    self.optimizer.zero_grad()
    loss.backward()

    if clip_error:
        for param in self.nn.parameters():
            param.grad.data.clamp_(-1,1)

    self.optimizer.step()

    if self.number_of_frames % update_target_frequency == 0:
        self.target_nn.load_state_dict(self.nn.state_dict())

    if self.number_of_frames % save_model_frequency == 0:
        save_model(self.nn)

    self.number_of_frames += 1
```

```python
class Atari(object):
    def __init__(self, Agent, label):
        self.name = label
        
        self.nn = Convolutional_Network().to(device)
        self.target_nn = Convolutional_Network().to(device)

        self.loss_func = nn.MSELoss()
        #self.loss_func = nn.SmoothL1Loss()
        
        self.optimizer = optim.Adam(params=self.nn.parameters(), lr=learning_rate)
        #self.optimizer = optim.RMSprop(params=mynn.parameters(), lr=learning_rate)
        
        self.number_of_frames = 0
        
        if resume_previous_training and os.path.exists(file2save):
            print("Loading previously saved model ... ")
            self.nn.load_state_dict(load_model())
        
setattr(Atari, "calculate_epsilon", atari_calculate_epsilon)
setattr(Atari, "select_action", atari_select_action)
setattr(Atari, "optimize", atari_optimize)
```

```python
memory = ExperienceReplay(replay_mem_size)

def atari_simulation(num_episodes, agent, label):
    memory = ExperienceReplay(replay_mem_size)
    actor = Agent.factory(agent, label)

    rewards_total = []

    frames_total = 0 
    solved_after = 0
    solved = False

    start_time = time()
    
    for i_episode in range(num_episodes):
    
        state = env.reset()

        score = 0
        #for step in range(100):
        while True:

            frames_total += 1

            epsilon = actor.calculate_epsilon(frames_total)

            #action = env.action_space.sample()
            action = actor.select_action(state, epsilon)

            new_state, reward, done, info = env.step(action)

            score += reward

            memory.push(state, action, new_state, reward, done)
            actor.optimize()

            state = new_state

            if done:
                rewards_total.append(score)

                mean_reward_100 = sum(rewards_total[-100:])/100

                if (mean_reward_100 > score_to_solve and solved == False):
                    print("SOLVED! After %i episodes " % i_episode)
                    solved_after = i_episode
                    solved = True

                if (i_episode % report_interval == 0 and i_episode > 0):

                    #plot_results()

                    print("\n*** Episode %i *** \
                          \nAv.reward: [last %i]: %.2f, [last 100]: %.2f, [all]: %.2f \
                          \nepsilon: %.2f, frames_total: %i" 
                      % 
                      ( i_episode,
                        report_interval,
                        sum(rewards_total[-report_interval:])/report_interval,
                        mean_reward_100,
                        sum(rewards_total)/len(rewards_total),
                        epsilon,
                        frames_total
                              ) 
                      )

                    elapsed_time = time() - start_time
                    print("Elapsed time: ", (elapsed_time))

                break


    print("\n\n\n\nAverage reward: %.2f" % (sum(rewards_total)/num_episodes))
    print("Average reward (last 100 episodes): %.2f" % (sum(rewards_total[-100:])/100))
    if solved:
        print("Solved after %i episodes" % solved_after)


    env.close()
    env.env.close()
```
