```python
# Problem #1
# If we list all the natural numbers below 10 that are multiples of 3 or 5, 
# we get 3, 5, 6 and 9. The sum of these multiples is 23. Find the sum of 
# all the multiples of 3 or 5 below 10000000 
start = time()

print ("Answer:\t ", sum(x for x in range(10000000) if (x % 3 == 0 or x % 5 == 0))) #The dirty way

print("Time:\t ", time() - start)
```

```python
# Problem 1A
# Find the sum of all the multiples integers of A or B below an integer limit L.
start = time()

def MSumTo(Int,Limit):
    stop = int((Limit-1)/Int)
    return(int(Int*stop*(stop + 1)/2)) # This uses arithmetic sum

def SumMulInt(A,B,Limit): return(MSumTo(A,Limit)+MSumTo(B,Limit)-MSumTo(A*B,Limit)) #More code, but much faster
    
print("Answer:\t ",SumMulInt(3,5,10000000))
print("Time:\t ", time() - start)
```

```python
# Problem 2
# By considering the terms in the Fibonacci sequence whose values do not 
# exceed four million, find the sum of the even-valued terms.

# Since the only time fibonacci's are even is when two of the proceeding values are odd...
start = time()

fib = [1,1]
def EvenFibSum(Limit):
    while fib[-1]+fib[-2] < Limit:
        fib.append(fib[-1]+fib[-2])
#    return(sum([x for x in fib if x%2 == 0])) # also works, but I want to cut out the test
    return(sum([x for x in fib[2::3]]))

print("Answer:\t ", EvenFibSum(4000000))
print("Time:\t ", time() - start)
```

```python
# Problem 2A 
# Instead of using 1 and 1 as the start of sequence,
# Let the pseudo-fibonacci sequence be defined with two starting seads A and B such that PF(n + 1) = PF(n) + PF(n-1).
# What values of A and B create the most primes inside of their PF sequence such that all pseudo-fibonacci sequences 
# whoes values do not exceed two million?
start = time()
Limit = 2000000

# Seive Algorithim

def primes(n):
    sieve = [True] * n
    for i in range(3,int(n**0.5+1),2):
        if sieve[i]:
            sieve[i*i::2*i]=[False]*(int((n-i*i-1)/(2*i)+1))
    return [2] + [i for i in range(3,n,2) if sieve[i]]

PrimesUnderLimit = primes(Limit)
SetPUnderLimit = set(PrimesUnderLimit) 

# Primes in standard fibonacci as minimum for test

def PrimeFib(A,B,L):
    fib = [A,B]
    while fib[-1]+fib[-2] < L:
        fib.append(fib[-1]+fib[-2])
    return([x for x in fib if x in SetPUnderLimit])

def FibPrint(A,B,L):
    fib = [A,B]
    while fib[-1]+fib[-2] < L:
        fib.append(fib[-1]+fib[-2])
    return(fib)

len(PrimeFib(1,1,Limit))

# To beat or tie the standard fibonacci we need at least 9 primes in list. All of the candidates on each PF
# list cannot be primes due to additive nature. No primes will ever be generated if both A and B are even.
# At best, only 2 out of every three PF list could be prime. Need to test a candidates 89A + 144B < Limit 
# if both A and B are odd, or 144A + 233B < Limit otherwise.

Candidates = []

for i in range(1,int(Limit/233)):
    for j in range(i,int(Limit/233)):
        if (i%2 == 1 or j%2 == 1):
            Candidates.append([i,j])
            
max_count = 0
            
for i in range(0,len(Candidates)):
    A = Candidates[i][0]
    B = Candidates[i][1]
    subject = len(PrimeFib(A,B,Limit))
    if subject >= max_count:
        Final = [FibPrint(A,B,Limit), subject ]
        max_count = subject

print("Answer:\t ", Final)
print("Time:\t ", time() - start)
```

```python
# Problem 3
# The prime factors of 13195 are 5, 7, 13 and 29.
# What is the largest prime factor of the number 317584931803 ?
start = time()

prime_factor = []
n = 317584931803
i=int(2)
while i<=n:
    if n % i == 0:
        n = n / i
        prime_factor.append(i)
        i = i - 1
    i=i+1
    
print("Answer:\t ", prime_factor)
print("Time:\t ", time() - start)
```

```python
# Problem 3A
# What are all the prime factors of the number 2^230 -1 ?
start = time()

def gcd(a,b):
    while a:
        a,b = b%a,a
    return b

def primes(n):
    sieve = [True] * n
    for i in range(3,int(n**0.5+1),2):
        if sieve[i]:
            sieve[i*i::2*i]=[False]*(int((n-i*i-1)/(2*i)+1))
    return [2] + [i for i in range(3,n,2) if sieve[i]]

PrimeList = primes(10000)


def Floyd(N, PrimeList):
        if N > 10000: Limit = 10000
        else: Limit = N
            
        for i in PrimeList:
            if N%i == 0: return i
            
        x = random.randint(Limit,N)
        y = x
        c = 2 #random.randint(Limit, N-1)
        g = 1
        while g==1:             
                x = ((x*x)%N+c)%N
                y = ((y*y)%N+c)%N
                y = ((y*y)%N+c)%N
                g = gcd(abs(x-y),N)
        
        return g

N = 2**230 - 1 
print(N)
Factors = list()

while N != 1:
    A = Floyd(N,PrimeList)
    N = N//A
    Factors.append(A)
    print(N, Factors)    

print("Time:\t ", time() - start)
```

```python
# Problem 4
# A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is
# 9009 = 91 × 99.
# Find the largest palindrome made from the product of two 3-digit numbers.

start =time()

print("Answer:\t ", max([i*j for i in range(999,100,-1) for j in range(999,100,-1) if str(i*j) == str(i*j)[::-1]]))
print("Time:\t ", time() - start)
```

```python
# Problem 4A
# An abundant number is a number for which the sum of its proper divisors is greater than the number itself (ex: 12)
# Find the largest palindrome under 100 million which is the product of two abundant numbers.

start =time()

def Abundant_Pal(x):
    for i in range(11,x):
        if str(i) == str(i)[::-1] and sum(divisors(i)) > i: yield i

def divisors(n):
    List = [1]
    largest = int(n**.5)

    # special-case square numbers to avoid yielding the same divisor twice
    if largest * largest == n:
        List.append(int(largest))
    else:
        largest += 1

    # all other divisors
    for i in range(2, largest):
        if n % i == 0:
            List.append(int(i))
            List.append(int(n / i))
    
    return sorted(List)

def is_abundant(n):
    if n < 12:
        return False
    return sum(divisors(n)) > n

Candidate = []
for i in Abundant_Pal(10000000):
    List = [x for x in divisors(i) if is_abundant(x)]
    for j in List:
        if i//j in List: 
            Candidate.append(i)

print("Answer:\t ", max(Candidate))
print("Time:\t ", time() - start)
```

```python
# Problem 5
# 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.
# What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?
start =time()

factors=[]
Answer=1
for n in range(1,21):
    for m in factors:
        if n%m == 0:
            n=n//m
    factors.append(n)
    Answer *= n

print("Answer:\t ", Answer) 
print("Time:\t ", time() - start)
```

```python
# Problem 6 
# Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.
start =time()

def Summo_Summer(size):
    return sum([x for x in range(size + 1)])**2 - sum([x**2 for x in range(size + 1)])

print("Answer:\t ", Summo_Summer(100))
print("Time:\t ", time() - start)
```

```python
# Problem 6A
# Let a Pythagorean square be a square number that is a sum of two other squares (ex 5^2 = 3^2 + 4^2). Find the 
# difference between the sum of the squares of the first hundred thousand Pythagorean squares and the square of the sum
# of the same first hundred thousand Pythagorean squares
start =time()

def Pythagorean_Square(limits) :
    Triple_List=set()
    c, m = 0, 2
    while c < limits :      
        for n in range(1, m) :
            a = m * m - n * n
            b = 2 * m * n
            c = m * m + n * n
            if c > limits :
                return (list((Triple_List)))
                break
            #only need the c part of the triple
            i = 1
            while i < limits//c:
                Triple_List.add(i*c)
                i += 1
 
        m = m + 1

limit = 100000

def List_SqSummer(List):
    return sum([x for x in List])**2 - sum([x**2 for x in List])

print("Answer:\t ", List_SqSummer(Pythagorean_Square(limit)))
print("Time:\t ", time() - start)
```

```python
# Visualize prime numbers as a "Klauber triangle"
# The maths behind this code is described in the scipython blog
# article at https://scipython.com/blog/the-klauber-triangle/
# Christian Hill, November 2016.

n = 100
ncols = 2*n+1
nmax = n**2

# Prime numbers up to and including n**2.
primes = np.array([n for n in range(2,n**2+1) if all(
                        (n % m) != 0 for m in range(2,int(np.sqrt(n))+1))])
a = np.zeros(nmax)
a[primes-1]=1

arr = np.zeros((n, ncols))
for i in range(n):
    arr[i,(n-i):(n+i+1)] = a[i**2:i**2+2*i+1]

fig, ax = plt.subplots()
ax.matshow(arr, cmap=cm.binary)
ax.axis('off')
# Ensure the Axes are centred in the figure
ax.set_position([1,1,8,8])
plt.savefig('klauber_triangle.png')
plt.show()
```
